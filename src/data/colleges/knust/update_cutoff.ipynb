```python
from bs4 import BeautifulSoup
import json
import os
```

```python

# Read the HTML file
with open('knustdata.json', 'r', encoding='utf-8') as f:
    data = f.read()
```

```python
# convert data to list
list_data = json.loads(data)
```

```python
for i, program in enumerate(list_data[1]):
    cutoff = program['cutoff point']["firstChoice"]
    print(f"{i+1}. {program['program']} - Cutoff: {cutoff}")
```

```python
for i, program in enumerate(list_data[1]):
    cutoff = program['cutoff point']["firstChoice"]
    new_cutoff =  {
        "firstChoiceSubjectArea": cutoff,
        "firstChoiceNonSubject": None,
        "fullFeePaying": None,
        "secondChoice": None,
        "subjectRequirements": None
    }
    program['cutoff point'] = new_cutoff

list_data
```

```python

# Create directory if it doesn't exist

# Save to JSON file
with open("knustdata.json", 'w', encoding='utf-8') as f:
    json.dump(list_data, f, indent=2)
```
